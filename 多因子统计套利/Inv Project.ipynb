```python
'''使用PCA降维35个宏观因子'''
def eigValPct(eigVals,percentage):
    '''通过方差的百分比来计算将数据降到多少维是比较合适的，
    函数传入的参数是特征值和百分比percentage，返回需要降到的维度数num'''
    sortArray=np.sort(eigVals) #使用numpy中的sort()对特征值按照从小到大排序
    sortArray=sortArray[-1::-1] #特征值从大到小排序
    arraySum=sum(sortArray) #数据全部的方差arraySum
    tempSum=0
    num=0
    for i in sortArray:     #取出特征值
        tempSum = tempSum + i      #特征值累计和
        num+=1
        print(i/arraySum)
        if tempSum>=arraySum*percentage:       #判断累计贡献率是否达到要求，此代码中这个值为0.88
            print(tempSum / arraySum)   #输出累计方差贡献率
            return num                  #返回达到要求所需要的维度数量

def pca(dataMat,percentage=0.9):
    '''pca函数有两个参数，其中dataMat是已经转换成矩阵matrix形式的数据集，列表示 特征；
    其中的percentage表示取前多少个特征需要达到的方差占比，默认为0.9'''
    meanVals=np.mean(dataMat,axis=0)  #对每一列求平均值，因为协方差的计算中需要减去均值
    meanRemoved=dataMat-meanVals
    covMat=np.cov(meanRemoved,rowvar=0)  #cov()
    eigVals,eigVects=np.linalg.eig(np.mat(covMat))  #利用numpy中寻找特征值和特征向量的模块linalg中的eig()方法
    k=eigValPct(eigVals,percentage) #要达到方差的百分比percentage，需要前k个向量
    print(k)

    eigValInd=np.argsort(eigVals)  #对特征值eigVals从小到大排序
    eigValInd=eigValInd[:-(k+1):-1] #排好序的特征值，从后往前取k个
    redEigVects=eigVects[:,eigValInd]   #返回排序后特征值对应的特征向量redEigVects（主成分）
    lowDDataMat=meanRemoved*redEigVects #将原始数据投影到主成分上得到新的低维数据lowDDataMat
    return lowDDataMat          #将低维度数据返回


pca_35 = np.loadtxt("PCA_35factor.csv",delimiter=',') 
lowDDataMat = pca(pca_35,0.88)    
np.savetxt('pca_7pc.csv',lowDDataMat,delimiter=',')   #结果导出
```

```python
'''匹配因子风险，选择8只股票tarck ETF50'''
data=pd.read_csv("dataset2.csv",encoding='gbk')
#导入数据到data
data=data.drop(columns='日期')
#删除日期列
coefs=np.zeros(shape=(43,7))#创建一个系数0矩阵
for i in range(43):#依次将指数收益率和个股收益率进行回归
    newdata=data.iloc[:,[i,43,44,45,46,47,48,49]].dropna()#取出自变量和因变量并删除空值
    y=newdata.iloc[:,0]#因变量
    X=newdata.iloc[:,1:]#自变量
    linreg = LinearRegression()  
    model=linreg.fit(X, y)#构建模型
    coefs[i]=linreg.coef_#将模型得到的系数放入系数矩阵中

coefs=pd.DataFrame(coefs,index=data.columns[0:43],columns=data.columns[43:])#
coefs.to_csv('系数矩阵2.csv',encoding='gbk')#导出系数矩阵
index=data.columns[1:43]
#ran_index=['招商银行','保利地产','恒瑞医药','贵州茅台','海螺水泥','中国平安','中国建筑','中国石油']#by industry
#ran_index=['中国平安','招商银行','贵州茅台','兴业银行','民生银行','中信证券','浦发银行','伊利股份']#by size
ran_index=['浦发银行','民生银行','宝钢股份','中国石化','南方航空','中信证券','招商银行','保利地产']#by random
#ran_index=np.random.choice(index,8)
ran_coefs=coefs.loc[ran_index,:].values.T#取出八只股票系数矩阵并转置
ones=np.ones(8)#构建系数为1的行
ran_coefs=np.insert(ran_coefs,0,values=ones).reshape(8,8)#构建系数矩阵
coef=coefs.iloc[0,:].values
one=np.ones(1)
coef=np.insert(coef,0,values=one) #构建常数项
x=np.linalg.solve(ran_coefs,coef)
weight=x.T
b = np.array([[0]])
Weight_ = np.insert(weight,0, values=b, axis=0)
np.savetxt("Weight.csv", Weight_, delimiter=',')
ran_index1 = ",".join(ran_index)
print("选取的股票为:"+ran_index1)
print("权重为：")
print(x)
```
